```python
import pickle 
import pandas as pd
```

```python
mape_knn_fk = pickle.load(open("mape-knn_fk.pkl", "rb"))

mape_knn_lstm_fk = pickle.load(open("mape-knn_lstm_fk.pkl", "rb"))

mape_knn_lstm = pickle.load(open("mape-knn_lstm.pkl", "rb"))

mape_knn_simple = pickle.load(open("mape-knn_simple.pkl", "rb"))

mape_knn_weight = pickle.load(open("mape-knn_weight.pkl", "rb"))

```

```python
mse_knn_fk = pickle.load(open("mse-knn_fk.pkl", "rb"))

mse_knn_lstm_fk = pickle.load(open("mse-knn_lstm_fk.pkl", "rb"))

mse_knn_lstm = pickle.load(open("mse-knn_lstm.pkl", "rb"))

mse_knn_simple = pickle.load(open("mse-knn_simple.pkl", "rb"))

mse_knn_weight = pickle.load(open("mse-knn_weight.pkl", "rb"))
```

```python
acc_knn_fk = []
acc_knn_lstm_fk = []
acc_knn_lstm = []
acc_knn_simple = []
acc_knn_weight = []

for i in mape_knn_fk:
    acc_knn_fk.append(1-i)##########
for i in mape_knn_lstm_fk:
    acc_knn_lstm_fk.append(1-i)############
for i in mape_knn_lstm:
    acc_knn_lstm.append(1-i)########
for i in mape_knn_simple:
    acc_knn_simple.append(1-i)#########
for i in mape_knn_weight:
    acc_knn_weight.append(1-i)#############
```

```python
import matplotlib.pyplot as plt
%matplotlib inline



e = range(0,len(mape_knn_fk))
print(e)
#et = Time[:20]
fig = plt.gcf()
fig.set_size_inches(20,10)
plt.plot(e[:20], acc_knn_fk[:20], label='acc_knn_fk' )
plt.plot(e[:20], acc_knn_lstm_fk[:20], label='acc_knn_lstm_fk' )
plt.plot(e[:20], acc_knn_lstm[:20], label='acc_knn_lstm' )
plt.plot(e[:20], acc_knn_simple[:20], label='acc_knn_simple' )
plt.plot(e[:20], acc_knn_weight[:20], label='acc_knn_weight' )


plt.xlabel('Time', fontsize=20)
plt.ylabel('ACCURACY', fontsize=20)
plt.title('Accuracy of Models')
plt.legend()
plt.figure()
plt.show()
```
